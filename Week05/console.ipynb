<h1> 1. Search for Signs Of Credentials Collection In The Registry </h1>

```python
# This first question will have commented code

import glob
import urlHaus as CSVer

# glob gets all the filenames into a list from a directory string.
files = glob.glob("../logs/w32processes/*")
listOfLsa = []

# Replaces all backslashes with forward slashes for Python's sake of finding the files.
for x in range(0, len(files)):
    files[x] = files[x].replace("\\","/")

# Setting the search terms for the python file.
searchTerm = ["HKLM\\SYSTEM\\CurrentControlSet\\Control\\Lsa"]

# Looped call of the python file so it looks in every file, then appends found matches to searchterm to listOfLsa
for z in range(0, len(files)):
    listOfLsa.append(CSVer.urlHaus(files[z], searchTerm)) 

# Cleans up nasty output of the above calls, in this case removing any None list entries.
for y in range(0, len(listOfLsa)):
    try:
        if listOfLsa[y] == None:
            listOfLsa.pop(y)
    except:
        pass

# Uses list comprehension to remove any empty lists in the list listOfLsa
listOfLsa = [x for x in listOfLsa if x]

# Prints out listOfLsa end results in a nice, easy-to-read manner.
for g in range(0, len(listOfLsa)):
    #print(listOfLsa[0][0])
    print("**********************************************************************")
    print("Type of Attack: Unsecured Credentials in Registry")
    print("Arguments: '"+ listOfLsa[g][0][1] + "' Hostname: '" + listOfLsa[g][0][2] + "' Name: '" + listOfLsa[g][0][3] + "' Path: '" + listOfLsa[g][0][4] + "' PID: '" + listOfLsa[g][0][5] + "' Username: '" + listOfLsa[g][0][6] + "'")

```
